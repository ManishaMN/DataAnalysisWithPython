```python
#summary statistics using .agg()
#computes custom summary stats , takes a function that will perform calculation of an argument
def pct95(col):
    return col.quantile(0.95)

```

```python
sales[['temperature_c', 'unemployment']].agg(pct95) #passing multiple columns for same statistics
```

```python
def pct75(col):
    return col.quantile(0.75)

sales['temperature_c'].agg([pct95,pct75]) #passing list of functions
```

```python
sales[['temperature_c', 'unemployment']].agg([pct95, pct75]) #passing multiple columns and list of functions
```

```python
#counting column values , method returns count of unique values in a column
sales['type'].value_counts()
```

```python
sales['type'].value_counts(sort=True) #will order counts highest to lowest if True
```

```python
#groupby, calculate summary statistics on different group rather than entire column
#groupby takes column name to groupby as argument, subset column to summarize in square brackets, call the summary statistic method.

sales.groupby('department')['type'].count()
```

```python
#similar to ungrouped statistics we can use .agg() to calcualte multiple summary statistics
sales.groupby('department')['temperature_c'].agg([max,min])
```

```python
#grouping by multiple varaibles ( pass a list of column names, subset values of interst and call a summary statistics)

sales.groupby(['department', 'type'])['temperature_c'].agg([max,min])
```

```python
#grouping by multiple columns and aggregating by multiple columns
sales.groupby(['department', 'type'])[['temperature_c','fuel_price_usd_per_l']].agg([max,min])
```

```python
#pivot tables
#calcualtes summary statistics similar to groupby , pass column to index argument, columnn to summarize to values
sales.pivot_table(values='temperature_c', index='type') #default is mean
```

```python
import numpy as np
sales.pivot_table(values='temperature_c', index='type', aggfunc = np.median) #to obatin median
```

```python
sales.pivot_table(values='temperature_c', index='type', aggfunc = [np.median, np.mean]) #to obtain median
```

```python
#pivoting on 2 variables ( first column to groupby to index and sceond column to groupby to columns)
sales.pivot_table(values='temperature_c', index='type', columns = 'is_holiday')
```

```python
#margins = true will give mean of values in columns and rows
sales.pivot_table(values='temperature_c', index='type', columns = 'is_holiday', margins = True)
```
